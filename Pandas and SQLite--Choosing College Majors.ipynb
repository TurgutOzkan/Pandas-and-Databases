# Aggregate Functions

These function provides summaries of a column. For example, you can query about the sum, average, or minimum of a column by calling these functions


```python
# The Major with the highest total number
pd.read_sql_query("SELECT Major, MAX(Total) FROM recent_grads", conn)
```

```python
# The Major with the lowest total number
pd.read_sql_query("SELECT Major, MIN(Total) FROM recent_grads", conn)
```

```python
# count how many major category are in our database
pd.read_sql_query("SELECT Count(Major_category) FROM recent_grads", conn)
```

```python
# The above includes duplicates too. To address that issue:
pd.read_sql_query("SELECT  Count(DISTINCT Major_category) FROM recent_grads", conn)
```

```python
# You can also see these 16 major categories by:
pd.read_sql_query("SELECT DISTINCT Major_category FROM recent_grads", conn)
```

```python
#  return unique pairings of the requested columns
pd.read_sql_query("SELECT DISTINCT Major, Major_category FROM recent_grads LIMIT 5",conn)
```

```python
#  Average Unemployment Rate by Major_category
pd.read_sql_query("SELECT AVG(Unemployment_rate), Major_category FROM recent_grads GROUP BY Major_category",conn)
```

```python
#  You can confirm this by checking average unemployment rate by education, which is the same number as above--0.051702
pd.read_sql_query("SELECT AVG(Unemployment_rate), Major_category FROM recent_grads WHERE Major_category='Education'",conn)
```

```python
pd.read_sql_query("SELECT Major_category, AVG(Unemployed) / AVG(Total) AS share_employed FROM recent_grads GROUP BY Major_category;", conn)
```

# Function for making queries

```python
def run_query(q):
    with sqlite3.connect(db) as conn:
        return pd.read_sql(q, conn)
```

```python
run_query("SELECT * FROM recent_grads WHERE Total >3000")
```

#Create a  function that returns a list of all tables in the database.

```python
def all_tables():
    q = '''
    SELECT
        name,
        type
    FROM sqlite_master
    WHERE type IN ("table","view");
    '''
    return run_query(q)

all_tables()
```

We have only one table and this is why it produced one row.

# Creating tables

```python
# We will create some madeup table and save into our database.
cur.execute("CREATE TABLE students (id integer, Name text, Address text, GPA float)")
conn.commit()
```

```python
# see if our new table is in the database.
all_tables()
```

One row is not ideal so let;s populate this table with more information. We will use fake dataset creator from https://github.com/joke2k/faker

```python
from faker import Faker
fake = Faker()
```

```python
#Creating fake 1000 students
name=[]
for i in range(0,1000):
    name.append(fake.name())
name[0:5]
```

```python
# Creating fake 1000 GPAs
gpa=[]
for i in range(0,1000):
    gpa.append(round(random.uniform(2.5, 4.0),2))
gpa[0:5]
```

```python
# Creating fake 1000 Addresses
addr=[]
for i in range(0,1000):
    addr.append(fake.address())
addr[0:5]
```

```python
# Creating 1000 student ID
ids=[]
for i in range(1000,2000):
    ids.append(i)
ids[0:5]
```

```python
data_tuples = list(zip(ids, name, addr, gpa))
st_df=pd.DataFrame(data_tuples, columns=['id', 'Name', 'Address', 'GPA'])
st_df.shape
```

Writing to our database using pandas

```python
st_df.to_sql("students", conn, if_exists="replace")
conn.commit()
```

```python
pd.read_sql_query("SELECT * FROM students", conn).head()
```

Lastly, we can add a column into database. We will use this column in JOIN operations

```python
#Let's get the major categories from student dataframe
maj_cat=df.Major_category.value_counts().index.tolist()
maj_cat
```

```python
for x in st_df.index.values:
    st_df.set_value(x,'Major_category',random.choice(maj_cat))
st_df.head()
```

```python
st_df.to_sql("students", conn, if_exists="replace")
conn.commit()
```
